##### c. Modifying DataFrame objcets

```python
# Mengubah nama kolom

data_titanic.rename(columns = {'survived': 'passenger_survived'})
```

```python
# Menambah kolom menggunakan append method
# Membuat DataFrame baru dengan variabel grup_umur

grup_umur=pd.DataFrame({'class 1':[22,40],'class 2':[24,50],'class 3':[20,45]})

# menambah kolom ke dataframe
grup_umur['class 4'] = [18,40]
grup_umur
```

```python
# menambahkan kolom dengan concat, misalkan dari data yang terpisah

class5 = pd.Series([31,48])
grup_umur = pd.concat ([grup_umur, class5], axis = 1) # axis menujukkan baris 1
grup_umur
```

```python
# Delete kolom menggunakan del()

del grup_umur[0]

print (grup_umur)
```

```python
# Delete kolom dengan drop()

grup_umur.drop([0], axis =1, inplace = True) 
grup_umur
```

##### e. Indexers and selection of subsets of data

* iloc: memilih data menggunakan posisi integer, mirip saat kita mengakses elemen di list dan array

```python
# menggunakan data_cuaca, menjadikan kolom Date sebagai index

data_cuaca = data_cuaca.set_index(['Date'])
```

##### h. Combine objects in Pandas

* Concat 


```python
# buat variabel periodic_table
periodic_table=pd.DataFrame({'Atomic Number':[1,2,3,4,5],'Element': 
                             ['Hydrogen','Helium','Lithium','Beryllium','Boron'],
                             'Symbol':['H','He','Li','Be','B']})
```

```python
periodic_table_extended=pd.DataFrame({'Atomic Number':[8,9,10],'Element':['Oxygen','Fluorine','Neon'],'Symbol':['O','F','Ne']})

#Join menggunakan pd.concat()
pd.concat([periodic_table,periodic_table_extended])
```

```python
# menggunakan inner join parameter
pd.concat([periodic_table,periodic_table_extended],axis=1,join='inner')
```

```python
# Merge

periodic_table=pd.DataFrame({'Atomic Number':[1,2,3,4,5],
                             'Element':['Hydrogen','Helium','Lithium','Beryllium','Boron'],
                             'Symbol':['H','He','Li','Be','B']})

periodic_table_extended=pd.DataFrame({'Atomic Number':[1,2,3],'Natural':'Yes'})

periodic_table.merge(periodic_table_extended)
```

#### Seaborn vs Matplotlib


```python
%matplotlib inline
# buat data menggunakan function dari NumPy untuk generate data
a = np.random.RandomState(0)
x = np.linspace (0, 10, 500)
y = np.cumsum(a.randn(500, 6), 0)
```

```python
# plot data dengan style matplotlib 
plt.plot(x, y)
plt.legend('ABCDEF', ncol = 2, loc ='upper left')
```

```python
import seaborn as sns
sns.set()
```

```python
# Plot data dengan menggunakan Seaborn style
plt.plot(x,y)
plt.legend('ABCDEF', ncol = 2, loc ='upper left')
```

```python
# Histogram di seaborn

data = np.random.multivariate_normal (
[0,0], [[5,2], [2,2]], size = 2000)
data = pd.DataFrame(data, columns = ['x', 'y'])

for col in 'xy':
    plt.hist(data[col], alpha = 0.5)
```

```python
# Untuk membuat smooth estimate pada distribusi menggunakan kernel density estimation
# Terapkan sns.kdeplot
pd.options.mode.chained_assignment = None #default = 'warn'

for col in 'xy':
    sns.kdeplot(data[col], shade =  True)
```

```python
# Plot distribusi
sns.distplot(data['x'])
sns.distplot(data['y'])
```
