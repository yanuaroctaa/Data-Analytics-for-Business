### IF Statement

> Contoh 1 - IF statement

kita memiliki nilai ujian dari siswa A dan siswa B, kita diminta untuk menentukan apakah dengan nilai ujian tersebut, siswa PASS atau FAILED.

siswa a = 80

siswa b = 65


*   Kategori PASS >= 70, dan di bawah 70 adalah FAILED
*   Maka kita perlu membuat code dengan kondisi di atas


```python
# Membuat variable

a = 80
b = 65

# if else statement

if a or b >= 70:
  print (a, ' telah PASS, selamat!')
  print (b, ' telah FAILED, segera ambil remidial')
```

*   umur kurang dari 12, harga masuk pengunjung adalah 50 (dalam ribu)

## Break 


>> Contoh 3

Dari suatu data structure berupa list, akan dilakukan iterasi menggunakan while loop Kombinasikan *while loop* dengan *if*. Misalkan kita tertarik dengan mencari siswa yang bernama Siska dari kumpulan daftar nama Siswa. Menggunakan urutan dari *list*, maka tampilkan seluruh nama data siswa sampai dengan memenuhi kondisi yaitu mendapatkan nama "Siska". Di dalam body *while loop*, *if-statement* akan membandingkan tiap item (nama siswa) dari *list* dengan nama "Siska". 

Jika kondisi menjadi True, maka pernyataan *break* akan di-*excecute* sehingga proses *loop* akan dihentikan. 


```python
# Buat variabel
siswa_data = ['Ajeng', 'Tya', 'Andre', 'Siska','Budi'] #buat list terkait siswa data
i = 0 # urutan dari nol

# Buat body of code dengan while loop dan if
while True:                            
  print (siswa_data[i])              
  if siswa_data[i] == 'Siska':
    print ('Berhasil menemukan nama Siska')
    break 
    print ('Setelah pernyataan break')
  i += 1  # update ke item selanjutnya                             
  print ('Searching for Siska') 
print ('Instruksi while loop selesai') 
```
