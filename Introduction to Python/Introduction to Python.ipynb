> ## 1. Python Arithmetic Operators

Membantu untuk melakukan operasi matematika seperti penambahan, pengurangan, perkalian, etc.

Jika x = 16, dan y = 4


```python
# Variable
x = 16
y = 4

# Addition
print ('addition: ', x+y)

# Substraction
print ('Substraction: ', x-y)

# Multiplication
print ('Multiplication: ', x*y)

# Division
print ('Division: ', x/y)

# Modulus
print ('Modulus: ', x%y)

# Exponent
print ('Exponent: ', x**y)

# Floor Division
print ('Floor Division: ', x//y)
```

> ## 2. Python Comparison Operators

Membadingkan kedua nilai operand dan menghasilkan True atau False jika kondisi dari keduanya *match*

Jika a = 20; b = 15

```python
# Variable
a = 20
b = 15

# ==
print ('result: ', a == b)

# !=
print ('result: ', a != b)

# >
print ('result: ', a > b)

# <
print ('result: ', a < b)

# >=
print ('result: ', a >= b)

# <=
print ('result: ', a <= b)
```

> ## 3. Python Assignment Operators

Digunakan untuk menetapkan (*assign*) beberapa *values* ke beberapa variable

Jika a = 15, b = 20, c = 0

```python
# Variable
a = 15
b = 20
c = 0

# Equal 
c = a + b
print (c)

# add AND 
c += a
print (c)

# Substract AND 
c -= a 
print (c)

# Multiply AND 
c *= a
print (c)
```

> ## 4. Python Bitwise Operators

Digunakan untuk menjalankan operasi bit, dan bilangan desimal akan dirubah menjadi bentuk format binary. 

a = 50 
b = 17

*in-built function* **bin()** di Python digunakan untuk mendapatkan representasi binary dalam bentuk bilangan Integer.

```python
a = 50 
b = 17

print ('a=', a, ':', bin (a), 'b=', b, ':', bin(b))
```

> ## 5. Python Logical Operator

Digunakan sebuah program untuk pengambilan keputusan berdasarkan beberapa kondisi. Setiap operand diamsusikan sebagai sebuah kondisi yang memberikan nilai True atau False.


```python
# variable 
i = 25

# Logical AND 
if i < 30 and i > 18:
  print ('Condition is fulfilled')
else:
  print ('Condition is not fulfilled')

# Logical OR 
if i > 30 or i < 18:
  print ('Condition is fulfilled')
else:
  print ('Condition is not fulfilled')
```

> ## 6. Python Membership Operator

Digunakan untuk memvalidasi keanggotaan dari sebuah value yang berurutan semisal pada *strings, lists, tuples*, dll. 

```python
# variable 
i = 40
j = 50
list_values = {10,20,30,40}

k = i/2

# Membership operator in 
if i in list_values:
  print ('i ada di dalam list')
else:
  print ('i tidak ada di dalam list')

# Membership operator not in 
if j not in list_values:
  print ('j tidak ada di dalam list')
else:
  print ('j ada di dalam list')
  
# Membership operator in 
if k in list_values:
  print ('k ada di dalam list')
else:
  print ('k tidak ada di dalam list')
```

> ## 7. Python Identity Operators

Digunakan untuk menentukan apakah sebuah nilai merupakan tipe atau *class* tertentu. Selain itu, tipe operator ini digunakan untuk menentukan jika tipe data tertentu mengandung beberapa variabel. 


```python
x = ['tea', 'coffee']
y = ['tea', 'coffee']
z = x

print (x is z)

print (x is y)

print (x == y)
```

> Sets


```python
fruits = ['apple', 'orange', 'apple', 'pear', 'orange', 'banana']
basket = set(fruits) #fungsi set akan menghapus elemen yang duplikat dan menyisakan elemen-elemen yang unique
print (basket) #untuk melihat isi dari set yang baru dibuat dengan nama variabel basket. Hasilnya adalah {'apple', 'pear', 'banana', 'orange'}
print ('orange' in basket) #fungsinya untuk memeriksa apakah orange ada di dalam set basket, hasilnya adalah TRUE yang mengindikasikan bahwa orange ada di dalam set
print('pineapple' in basket) #hasilnya adalah FALSE, tidak ada pineapple di dalam set
```

```python
a = set('aioeueoiaeaeiou') #fungsinya untuk membuat set tanpa ada duplikat
b = set('bcokcbzo') #fungsinya untuk membuat set tanpa ada duplikat

print(a) # {'o', 'u', 'a', 'e', 'i'}
print(b) # {'z', 'o', 'c', 'k', 'b'}

# kita coba gunakan bitwise operator 

print(a&b) #akan menampilkan irisan huruf yang ada di kedua set a & b

print (a|b) # akan menampilkan semua huruf baik di set a maupun di set b

print(a-b) # akan menampilkan huruf yang ada di set a & tidak mengikutsertakan huruf yang beririsan dengan set b
```

> Dictionaries

```python
capitals = { 'AP' : 'Hyderabad', 'MH' : 'Mumbai'}
capitals['TN']= 'Chennai'

print (capitals['AP']) #menampilkan value dari AP yang ada di dalam dictionary

del capitals['TN'] #menghapus TN dari dalam dictionary

capitals['UP']= 'Luck now' #menambahkan UP kedalam dictionary

print ('AP' in capitals) #Memeriksa apakah key AP ada di dalam dictionary
print ('TN' in capitals) #Memeriksa apakah key TN ada di dalam dictionary, karena sudah dihapus maka hasilnya FALSE
```
